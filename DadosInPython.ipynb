```python
# Step 0 -> Understand the challenge you want to solve
# Step 1 -> Browse all files in the database -> Vendas folder
import os
import pandas as pd

lista_arquivo = os.listdir("/content/drive/MyDrive/PythonCourse/Vendas")
display (lista_arquivo)
```

```python
finalTable = pd.DataFrame()

# Step 2 -> Import sales databases

for arquivo in lista_arquivo:
  if "Vendas" in arquivo:
    table = pd.read_csv(f"/content/drive/MyDrive/PythonCourse/Vendas/{arquivo}")
    finalTable = pd.concat([finalTable, table], ignore_index=True)
display(finalTable)
```

```python
# Step 4 -> Calculate the best-selling product (in quantity)
productsTable = finalTable.groupby("Produto").sum()
productsTable = productsTable[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
display(productsTable)
```

```python
# Step 5 -> Calculate the most invoiced product (in revenue)
finalTable['Faturamento'] = finalTable['Quantidade Vendida'] * finalTable['Preco Unitario']
FatTable = finalTable.groupby("Produto").sum()
FatTable = FatTable[["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(FatTable)
```

```python
# Step 6 -> Calculate the store/city that sold the most (in revenue) - create a graph/dashboard
cityTable = finalTable.groupby('Loja').sum()
cityTable = cityTable[["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(cityTable)
```

```python
import plotly.express as px
fig = px.bar(cityTable, x=cityTable.index, y='Faturamento')
fig.show()
```
